```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.model_selection import KFold
```

```python
df=pd.read_csv(r'C:\Users\yatharth bansal\Downloads\nba_2013.csv')
```

```python
df.head()
```

Removing useless columns.

```python
df.drop(columns=['player','pos','bref_team_id','season','season_end'],inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
df[['x2p', 'x2pa', 'x2p.', 'efg.', 'ft', 'fta']].describe()
```

Removing missing values.

```python
df['fg.']=df['fg.'].fillna(df['fg.'].mean())
df['x3p.']=df['x3p.'].fillna(df['x3p.'].mean())
df['ft.']=df['ft.'].fillna(df['ft.'].mean())
df['x2p.']=df['x2p.'].fillna(df['x2p.'].mean())
df['efg.']=df['efg.'].fillna(df['efg.'].mean())
```

```python
df.describe()
```

Dividing our dataset.

```python
X=df.drop(columns='pts')
y=df['pts']
```

Scaling our dataset.

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=.30,random_state=355)
```

Creating a model of knn regressor.

```python
knn=KNeighborsRegressor()
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

Using grid search CV for tuning our model. 

```python
param_grid = { 'algorithm' : ['ball_tree', 'kd_tree', 'brute'],
               'leaf_size' : [25,27,30,32,34],
               'n_neighbors' : [3,5,7,9,10]
              }
```

```python
gridsearch = GridSearchCV(knn, param_grid,verbose=3)
```

```python
gridsearch.fit(X_train,y_train)
```

```python
gridsearch.best_params_
```

```python
knn=KNeighborsRegressor(algorithm= 'ball_tree', leaf_size= 25, n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```
